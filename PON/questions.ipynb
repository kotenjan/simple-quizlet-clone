Question:




Dokaž, že v QR rozkladu je j-ty sloupec matice X je lineární kombinací 0..j tych sloupců matice Q. 

Answer:
 
Důkaz:

Necht $X$ je matice a $QR$ je jeji rozklad, kde $Q$ je ortogonalni matice a $R$ je horni trojuhelnikova matice.

Pro j-ty sloupec matice $X$, oznaceny $X_{:j}$, QR rozklad implikuje:
$$X_{:j} = Q \left[ \begin{array}{c} R_{1j} \\ R_{2j} \\ \vdots \\ R_{jj} \\ 0 \\ \vdots \\ 0 \end{array} \right]$$

Protoze $R$ je horni trojuhelnikova, vsechny prvky pod diagonalou jsou nulove, konkretně vsechny prvky $R_{ij}$ pro $i > j$ jsou nulove.

Sloupec $X_{:j}$ lze napsat jako linearni kombinaci sloupcu $Q$:
$$X_{:j} = R_{1j}Q_{:1} + R_{2j}Q_{:2} + \ldots + R_{jj}Q_{:j}$$

Zde $Q_{:i}$ oznacuje i-ty sloupec $Q$.

Takze $X_{:j}$ je opravdu linearni kombinaci prvni j sloupcu $Q$, konkretně sloupcu 0 az j (pokud uvazujeme 0-indexaci). To dokazuje, ze j-ty sloupec matice $X$ je linearni kombinaci 0 az j-tych sloupcu matice $Q$.

$B, A, P$ jsou matice, dokaz, ze $B$ a $A$ maji stejna vlastni cisla pokud plati: $B=PAP^{-1}$ 

Answer:
 
Otázka: B, A, P jsou matice, dokaž, že B a A mají stejná vlastní čísla pokud platí: B = PAP^{-1}

Důkaz:

Vlastní čísla matice jsou řešeními charakteristické rovnice matice, tedy det(A - λI) = 0, kde I je jednotková matice a λ jsou vlastní čísla.

Vezmeme charakteristickou rovnici pro matici B:
det(B - λI) = 0

Dosadíme za B z daného vztahu:
det(PAP^{-1} - λI) = 0

Protože maticové násobení je distributivní, platí:
det(PAP^{-1} - PλP^{-1}) = det(P(A - λI)P^{-1})

Využijeme vlastnosti determinantu součinu matic:
det(P(A - λI)P^{-1}) = det(P) det(A - λI) det(P^{-1})

Determinant inverzní matice je inverzní hodnotou determinantu původní matice, takže:
det(P) det(P^{-1}) = 1/det(P) det(P) = 1

Tím dostáváme:
det(A - λI) = 0

To znamená, že λ, které jsou vlastními čísly matice A, jsou také vlastními čísly matice B, což bylo potřeba dokázat.

Question:




Kdy je matice symetrická? Z jaké množiny čísel jsou vlastní čísla sym. matice? Dokaž to. 

Answer:
 
když $X^T=X$, což implikuje, že musí být čtvercová.
vl. čísla jsou z $\mathbb{R}$. Což je dobrý, jelikož můžou být z $\mathbb{C}$
Matice je symetrická, pokud $A = A^T$, tedy pokud je rovna své transponované matici, což znamená, že musí být čtvercová.

Vlastní čísla symetrické matice jsou vždy reálná čísla. To lze dokázat využitím vlastností vlastních čísel a vlastních vektorů.

Pro vlastní vektor $x$ symetrické matice $A$ a jeho vlastní číslo $\lambda$, které může být komplexní, platí $Ax = \lambda x$. Když toto vynásobíme zleva transponovaným vlastním vektorem $x^T$, dostaneme $x^T A x = \lambda x^T x$. Protože $A$ je symetrická, můžeme napsat $x^T A x = x^T A^T x = (Ax)^T x = (\lambda x)^T x = \lambda^* x^T x$, kde $\lambda^*$ je komplexně sdružené číslo k $\lambda$.

Porovnáním obou stran dostaneme, že $\lambda x^T x = \lambda^* x^T x$, což implikuje, že $\lambda = \lambda^*$, a tedy vlastní číslo $\lambda$ musí být reálné, protože komplexně sdružené číslo je rovno původnímu číslu pouze pokud je to číslo reálné.

Question:




Armijova podmínka 

Answer:
 
Používá se v kontextu hledání vhodné délky kroku připotimalizaci
definujme pomocnou funkci $\phi(\alpha)$, co mi vrátí hodnotu $f$ pro další krok z bodu $x_k$  ve směru $p_k$, s délkou kroku $\alpha$
$\phi(\alpha) = f(x_k + \alpha p_k), \alpha \geq 0$
Armijova podmínka potom vypadá takto:
$\phi(\alpha_k) \leq \phi(0) + \alpha_kc\phi'(0)$
kde $0 < c < 1$
Tedy funkční hodnota má ležet pod zvolenou přímkou, nalezený krok by však neměl být příliš malý.
<!--ID: 1644142857952-->

Question:




Goldsteinova podmínka 

Answer:
 
Používá se v kontextu hledání vhodné délky kroku připotimalizaci
definujme pomocnou funkci $\phi(\alpha)$, co mi vrátí hodnotu $f$ pro další krok z bodu $x_k$  ve směru $p_k$, s délkou kroku $\alpha$
$\phi(\alpha) = f(x_k + \alpha p_k), \alpha \geq 0$
Goldsteinova podmínka potom vypadá takto:
$\phi(0) + \alpha_k (1-c)\phi'(0) \leq \phi(\alpha_k) \leq \phi(0) + \alpha_kc\phi'(0)$
kde $0 < c < \frac{1}{2}$
Tedy funkční hodnota má ležet mezi zvolenými přímkami: omezuje se nevhodná možnost, že nalezenýkrok bude příliš malý
Nevýhoda: můžeme minout optimálního hodnotu pro $\alpha_k$

Question:




Dokaž: $f:\Omega \to \mathbb{R}$ je spojitě diferencovatelná a $\Omega$ je konvexní. Pak platí, že $f$ je konvexní právě tehdy, když: $\forall x, y \in \Omega, f(y) \geq f(x) + \nabla f(x)^T \cdot (y-x)$ 

Answer:
 
Důkaz konvexity funkce spočívá ve splnění následující podmínky:

Pro spojitě diferencovatelnou funkci $f: \Omega \to \mathbb{R}$, kde $\Omega$ je konvexní množina, je funkce $f$ konvexní, pokud a jen pokud pro všechna $x, y \in \Omega$ platí, že:

$f(y) \geq f(x) + \nabla f(x)^T \cdot (y-x)$

Intuitivně to znamená, že hodnota funkce $f$ v bodě $y$ je vždy vyšší nebo rovna hodnotě funkce v bodě $x$ plus skalární součin gradientu funkce v bodě $x$ a vektoru směru $(y-x)$.

Jinými slovy, hodnota funkce na libovolném bodě $y$ není nižší než hodnota lineární aproximace funkce v bodě $x$, která je daná tečnou funkcí v $x$. Konvexnost zajišťuje, že graf funkce leží nad tečnou rovinou.

Ryzí konvexnost jde ještě dále, říká, že pokud $x \neq y$, pak nerovnost bude striktní, což znamená:

$f(y) > f(x) + \nabla f(x)^T \cdot (y-x)$

To znamená, že graf funkce je striktně nad tečnou rovinou, když $x \neq y$, a neleží přímo na ní. To znamená, že není žádná jiná bodová přímka mezi jakýmikoli dvěma body na grafu funkce.

pokud navíc platí $x \neq y$, tak $f(y) > f(x) + \nabla f(x)^T \cdot (y-x)$ ($f$ je ryze konvexní)

Question:




Popiš princip algoritmu pro QR rozklad. 

Answer:
 
Algoritmus QR rozkladu využívá faktu, že násobení ortogonálních matic vede k ortogonální matici a že inverze ortogonální matice je její transpozice. Pro každou pozici v R, kde má být nula, vyrobíme ortogonální matici Qi, kterou když vynásobíme X, vytvoříme v R nulu.

Postup je následující:

1. Vynásobíme matici X ortogonální maticí Q1, která vytváří nulu na požadované pozici prvního sloupce matice R.
2. Tento krok opakujeme pro další sloupce, kde každé Qi se zaměřuje na nulování prvků ve svém sloupci v R.

Jaký vztah je mezi vlastními čísly, jejich násobnostmi a diagonalizovatelnými maticemi? 

Answer:
 
Vztah mezi vlastními čísly, jejich násobnostmi a diagonalizovatelností matic je následující:

(i) Geometrická násobnost vlastního čísla (dimenze vlastního podprostoru) nikdy není větší než jeho algebraická násobnost (násobnost v charakteristickém polynomu).

(ii) Matice $A$ je diagonalizovatelná právě tehdy, když pro všechna vlastní čísla platí, že jejich geometrická násobnost je rovna algebraické násobnosti.

Pokud jsou geometrické násobnosti rovny algebraickým násobnostem pro odpovídající vlastní čísla.

Question:




Dokaž nebo vyvrať následující tvrzení: $h(X^TX) = h(XX^T) = h(X)$ pro každou matici $X \in \mathbb{R}^{m, n}$ 

Answer:
 
Matice $A$ je symetrická, pokud $A = A^T$, což znamená, že musí být čtvercová. Vlastní čísla symetrické matice jsou vždy z množiny reálných čísel, což je důležité, protože vlastní čísla matice obecně mohou být komplexní.

Důkaz:

1. Nechť $\lambda$ je vlastní číslo symetrické matice $A$ a $x$ je odpovídající vlastní vektor, takže $Ax = \lambda x$. 
2. Transponujeme obě strany rovnice: $(Ax)^T = (\lambda x)^T$ což dává $x^T A^T = \lambda^* x^T$, kde $\lambda^*$ je komplexně sdružené číslo k $\lambda$. 
3. Vzhledem k symetrii $A$, je $A^T = A$, a tedy $x^T A = \lambda^* x^T$. 
4. Násobíme $x$ zprava: $x^T Ax = \lambda^* x^T x$, což znamená, že $\lambda x^T x = \lambda^* x^T x$. 
5. Protože $x^T x$ je vždy reálné a nenulové pro ne-nulový vektor $x$, musí být $\lambda = \lambda^*$, což znamená, že $\lambda$ musí být reálné. 

To ukazuje, že i když definice vlastního čísla dovoluje komplexní čísla, v případě symetrické matice jsou vlastní čísla vždy reálná.

$x$ je OG ke všem řádkům $X$
$\Rightarrow Xx=\theta$
$\Rightarrow X^TXx = \theta$
$\Rightarrow x^TX^TXx = \theta$
$x^TX^TXx = (Xx)^T(Xx) = ||Xx||^2=\theta$
-> řádky $X$ a $X^TX$ generují stejný prostor, jinak by existoval $x$ z jednoho obalu, co by byl OG k druhému obalu. No když mají matice stejné lin. obal řádků, tak mají stejnou hodnost.

Question:




Dokaž nebo vyvrať, že je matice $X^TX$ pozitivně definitní ($X \in \mathbb{R}^{m, n}$) 

Answer:
 
v důkazu pro $h(X^TX) = h(XX^T) = h(X)$ jsme si ukázali, že platí
$\boldsymbol{x}^{T} \mathbf{X}^{T} \mathbf{X} \boldsymbol{x}=(\mathbf{X} \boldsymbol{x})^{T}(\mathbf{X} \boldsymbol{x})=\|\mathbf{X} \boldsymbol{x}\|^{2} \geq 0$ no a to znamená, že pro všechny $X^TX$ **neplatí**, že jsou vždy pozitivně definitní, ale že jsou vždy pozitivně semidefinitní.

Question:




Dej mi nějaký vlastnosti symetrické matice $S$ 

Answer:
 
* $S$ je diagonalizovatelná, její vlastní vektory lze volit tak, že tvoří ortonormální bázi.
* Existuje rozklad $S=QDQ^T$
* Všechna vlastní čísla matice $S$ jsou reálná
* Je-li $S$ pozitivně semidefinitní, jsou její vl. čísla nezáporná
* Je-li $S$ pozitivně definitní, jsou její vl. čísla kladná
* $h(s) = h(S^TS) = h(SS^T)$
* $S^TS$ a $SS^T$ jsou symetrické
* $S^TS$ a $SS^T$ jsou pozitivně semidefinitní
* $S^TS \in \mathbb{R}^{n, n}$ jsou pozitivně definitní $\iff$ $h(S) = n$.
* $SS^T \in \mathbb{R}^{m, m}$ jsou pozitivně definitní $\iff$ $h(S) = m$.
